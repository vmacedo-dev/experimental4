```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def incidencia(e0, mu, x):
    e = e0*np.exp(mu*x)
    return e

def mu(E, E0, x):
    mu = []
    valor = 0
    for i in range(6):
        calc = np.log(E[i+1]/E0)/x[i+1]
        mu.append(calc)
        valor += mu[i]
        i += 1
    return valor
        
        
```
