```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def incidencia(e0, mu, x):
    e = e0*np.exp(mu*x)
    return e

def mu(E, E0, x):
    mu = []
    valor = 0
    for i in range(6):
        calc = np.log(E[i+1]/E0)/x[i+1]
        mu.append(calc)
        valor += mu[i]
        i += 1
    return valor/6    
```

#### Calculando o valor de $\mu$

$$E = E_0e^{\mu x}$$

$$e^{\mu x} = \frac{E}{E_0}$$

$$\mu x = ln(\frac{E}{E_0})$$

$$\mu = \frac{ln(\frac{E}{E_0})}{x}$$

Com $E_0 = 14.69$ e o valor de $\mu$ medio podendo ser calculado por

$$\mu_{medio} = \frac{\sum\limits_{i=1}^{\mbox{}r}\mu_i}{i}$$

no nosso caso $r = 6$

```python
x = np.linspace(0, 100)
E0 = 14.69
x1 = [0, 10, 20, 30, 40, 60, 80]
y = [14.69, 4.39, 1.82, 1.42, 0.602, 0.411, 0.354] #medidas experimentais
u = mu(y, E0, x1)
print(u)
E = incidencia(E0, u, x)
```

```python
plt.plot(x, E, color='black', label='Ajuste de Beer-Lambert')
plt.scatter(x1, y, color ='red', label='Dados Experimentais')
plt.legend()
plt.xlabel('x(cm)')
plt.ylabel('E(KLx)')
#plt.grid()
plt.show()
```
